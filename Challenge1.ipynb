```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```

```python
print(iris_df.head())
print(iris_df.describe())
print(iris_df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(iris_df, hue='target')
plt.show()
```

```python
iris_df.to_csv('iris_dataset.csv', index=False)
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nvidia_data = yf.download('NVDA', start='2020-01-01', end='2025-01-01')
```

```python
print(nvidia_data.head())
print(nvidia_data.describe())
print(nvidia_data.info())
```

```python
plt.figure(figsize=(10,6))
nvidia_data['Close'].hist(bins=50, color='blue', alpha=0.7)
plt.title('Distribution of NVIDIA Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()
```

```python
nvidia_data['Daily Return'] = nvidia_data['Close'].pct_change()
plt.figure(figsize=(10,6))
nvidia_data['Daily Return'].plot(color='orange')
plt.title('NVIDIA Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(nvidia_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for NVIDIA Stock Data')
plt.show()
```
